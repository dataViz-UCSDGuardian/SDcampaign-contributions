```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df16 = pd.read_csv("financial_support_2016_datasd.csv")
df15 = pd.read_csv("financial_support_2015_datasd.csv")
```

```python
df16 = df16.drop(['report_period_from', 'report_period_to', 'intermediary_first', 'intermediary_last', 'intermediary_city', 'intermediary_state', 'intermediary_zip'], axis=1)
df15 = df15.drop(['report_period_from', 'report_period_to', 'intermediary_first', 'intermediary_last', 'intermediary_city', 'intermediary_state', 'intermediary_zip'], axis=1)
```

```python
df16['contributor_occ'].replace('', np.nan, inplace=True)
df16['contributor_occ'].replace(' ', np.nan, inplace=True)
df15['contributor_occ'].replace('', np.nan, inplace=True)
df15['contributor_occ'].replace(' ', np.nan, inplace=True)
df16.dropna(subset=['contributor_occ'], inplace=True)
df15.dropna(subset=['contributor_occ'], inplace=True)
```

```python
df_combined = pd.concat([df16, df15])
df_combined
```

```python
df_combined['council'] = df_combined['recipient_name'].str.lower()
df_combined = df_combined[df_combined['council'].str.contains('council')]
df_combined = df_combined.drop(['council'], axis=1)
```

```python
df_combined.groupby(['recipient_id', 'recipient_name']).size().sort_values(ascending=False)
```

```python
df_combined['contributor_occ'] = df_combined['contributor_occ'].str.title()
df_combined
```

```python
df_by_id = df_combined.groupby(['recipient_id', 'contributor_occ']).size().reset_index(name='Count')
```
